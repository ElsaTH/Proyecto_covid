```python
df_json = pd.read_json("covidgrupD.json") 
df_json
```

```python
df_json.reset_index(inplace=True)
```

```python
df_json
```

```python
fig = go.Figure(data=go.Scatter(x=df_json['index'],
                                y=df_json['t_d_averages'],
                                mode='lines')) # hover text goes here
fig.update_layout(title='Number of COVID cases over time',xaxis_title="Date",yaxis_title="Number of cases")
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Box(y=df_json.t_d_averages,
                     boxmean=True, # only mean visible on plot
                     marker_color="green",
                     name="T_D"))
fig.add_trace(go.Box(y=df_json.index,
                     boxmean='sd', # mean and SD visible on plot
                     marker_color="darkorchid",
                     name="index"))
fig.update_layout(title="Date&T_D")
fig.show() 
```
